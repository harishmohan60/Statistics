```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from scipy import stats

import os
print(os.getcwd())
```

```python
load_df = pd.read_csv("RawData/Accepted_Loan_Summary.csv")
load_df = load_df[['fico_score', 'loan_total', 'default_total', 'dti', 'avg_loan_amnt', 'median_annual_inc', 'avg_annual_inc', 'int_rate', 'default_rate', 'actual_default_rate']]
load_df.head()
```

```python
load_df.describe()
```

```python
load_df.cov()
```

```python
load_df.corr(method='spearman')
```

```python
load_df = load_df.dropna(subset=['median_annual_inc'])
sns.distplot(load_df['median_annual_inc'], hist=True)
```

```python
load_df[['fico_score','median_annual_inc']].corr('spearman')
```

```python
sns.regplot(x='fico_score', y='median_annual_inc', data=load_df)
```

```python
load_df[['fico_score','avg_annual_inc']].corr('spearman')
```

```python
sns.regplot(x='fico_score',y='avg_annual_inc',data=load_df)
```

```python
load_df[['fico_score','dti']].corr('spearman')
```

```python
sns.regplot(x='fico_score',y='dti',data=load_df)
```

```python
load_df[['fico_score','int_rate']].corr()
```

```python
sns.regplot(x='fico_score',y='int_rate',data=load_df)
```

```python
sns.distplot(load_df['int_rate'], hist=True)
```

```python
stats.probplot(load_df['int_rate'], dist="norm", plot=plt)
```

```python
load_df[['fico_score','loan_total']].corr('spearman')
```

```python
sns.regplot(x='fico_score',y='loan_total',data=load_df)
```

```python
coeff = np.poly1d(np.polyfit(load_df['fico_score'], load_df['loan_total'], 5))

t = load_df['fico_score']
plt.plot(load_df['fico_score'], load_df['loan_total'], 'o', t, coeff(t), '-')
plt.show()
```
